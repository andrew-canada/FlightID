- First, we graph all the flight data, highlight (red-green) based on the max-min of the predicted difference by the model.  

```python
import pydeck as pdk
import pandas as pd
import json
import numpy as np

#loading datasets
coordinates = json.load(open("../data/airport_coordinates.json",'r'))
df = pd.read_csv("../outputs/inflated_flights.csv")



df["src_lat"] = [coordinates[i][0] for i in df["city1"]]
df["src_lon"] = [coordinates[i][1] for i in df["city1"]]
df["dst_lat"] = [coordinates[i][0] for i in df["city2"]]
df["dst_lon"] = [coordinates[i][1] for i in df["city2"]]
df["pretty_dif"] = [round(i,2) for i in df["pct_diff"]]

# print(df.head())
avg = np.average(df["pct_diff"])
# print(df)
layers = []
for row in df.to_dict(orient="records"):
    # highlight = avg<row["pct_diff"]
    # act_dif =row["pct_diff"] * (-1 if row["actual_fare"] < row["predicted_fare"] else 1)
    layer = pdk.Layer(
        "ArcLayer",
        data=[row],
        get_source_position=["src_lon", "src_lat"],
        get_target_position=["dst_lon", "dst_lat"],
        get_source_color=[70 +(70000* max(-(row["pct_diff"]-avg),0)), 70+(5000* max((row["pct_diff"]-avg),0)), 20], # red ish if the fare is expensive green if it is chaper than the avg model difference
        get_target_color=[150, 150, 150],   # cyan at destination
        get_width=1.2,
        pickable=True,
        auto_highlight=True,
    )
    layers.append(layer)

# print(layers)


view_state = pdk.ViewState(
    latitude=40.0,
    longitude=-100.0,
    zoom=3.5,
    pitch=60,
    bearing=20,
)
r = pdk.Deck(
    layers=layers,
    initial_view_state=view_state,
    map_style="https://basemaps.cartocdn.com/gl/dark-matter-gl-style/style.json",
    tooltip={"html": "{city1} → {city2}\nAvg. Fare: {actual_fare} \n Difference from Predicted(ratio): {pretty_dif}"},                       # free, no API key needed
)

# Map of predicted model differences from actual fare prices
r
```

```python
import math

def is_long_distance(lat1, lon1, lat2, lon2, threshold_km=2000):
    """
    Returns True if the flight distance between two coordinates exceeds threshold.
    Default threshold: 1500 km (~930 miles) — roughly NYC to Denver.
    """
    R = 6371  # Earth radius in km

    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    distance_km = 2 * R * math.asin(math.sqrt(a))

    return distance_km >= threshold_km, round(distance_km)

layers=[]

for row in df.to_dict(orient="records"):
    # highlight = avg<row["pct_diff"]
    # act_dif =row["pct_diff"] * (-1 if row["actual_fare"] < row["predicted_fare"] else 1)
    check, dist = is_long_distance(row["src_lat"],
                            row["src_lon"],
                            row["dst_lat"],
                            row["dst_lon"])
    print(dist)
    if not check:
        continue
    layer = pdk.Layer(
        "ArcLayer",
        data=[row],
        get_source_position=["src_lon", "src_lat"],
        get_target_position=["dst_lon", "dst_lat"],
        get_source_color=[70 +(70000* max(-(row["pct_diff"]-avg),0)), 70+(5000* max((row["pct_diff"]-avg),0)), 20],   # pink at origin
        get_target_color=[150+(20000* max(-(row["pct_diff"]-avg),0)), 150+(1000* max((row["pct_diff"]-avg),0)), 150],   # cyan at destination
        get_width=1.2,
        pickable=True,
        auto_highlight=True,
    )
    layers.append(layer)

v = pdk.Deck(
    layers=layers,
    initial_view_state=view_state,
    map_style="https://basemaps.cartocdn.com/gl/dark-matter-gl-style/style.json",
    tooltip={"html": "{city1} → {city2}\nAvg. Fare: {actual_fare} \n Difference from Predicted(ratio): {pretty_dif}"},                       # free, no API key needed
)

# Map of predicted model differences from actual fare prices
v
```
